```python
# Filling missing values for numerical columns with median
for i in train.select_dtypes(include=['number']).columns:
    train.loc[:, i] = train[i].fillna(train[i].median())

# Filling missing values for categorical columns with mode
for i in train.select_dtypes(include=['object']).columns:
    train.loc[:, i] = train[i].fillna(train[i].mode()[0])
```

**6. Categorical column Analysis**

- Frequency table
- Relative Frequency
- Bar chart
- Pie chart

```python
# 1. Frequency Table
for i in cat_cols:
    keys=train[i].value_counts().keys()
    values=train[i].value_counts().values
    data=pd.DataFrame(zip(keys,values),columns=['Labels','Count'])
    print(data)
```

```python
# 2. Relative frequency table
for i in cat_cols:
    relative=train[i].value_counts(normalize=True)
    data1=pd.DataFrame(zip(train[i].value_counts().keys(),train[i].value_counts().values,relative),columns=['Labels','Counts','relative_frequency'])
    print(data1)
```

```python
# 3. Bar Chart 
for i in cat_cols:
    keys=train[i].value_counts().keys()
    values=train[i].value_counts().values
    plt.figure(figsize=(6,6))
    plt.bar(keys,values)
    plt.xlabel('Labels')
    plt.ylabel('Count')
    plt.title(i)
    plt.show()
```

```python
# 4. Pie Chart
for i in cat_cols:
    keys=train[i].value_counts().keys()
    values=train[i].value_counts().values
    plt.pie(values,labels=keys,autopct='%0.1f%%')
    plt.title(f'{i}')
    plt.show()
```

**7. Numerical column analysis**
- two ways
    - statical
    - data distribution  

```python
# 1.Statical 
l=[]  
index=['Count', 'Min', 'Max', 'Mean', 'Median', 'Mode', 'Std', '25p', '50p', '75p']  
for i in num_cols:    
    count=train[i].count()
    minn=train[i].min()
    maxx=train[i].max()
    mean=train[i].mean()
    median=train[i].median()
    mode=train[i].mode()[0] if not train[i].mode().empty else np.nan  
    std=train[i].std()
    _25p=np.percentile(train[i].dropna(), 25)  
    _50p=np.percentile(train[i].dropna(), 50)
    _75p=np.percentile(train[i].dropna(), 75)
    data=[count, minn, maxx, mean, median, mode, std, _25p, _50p, _75p]
    l.append(data)
df=pd.DataFrame(l,index=num_cols,columns=index)
df
```

**8.Outlire Analysis**
- Box plot.


```python
for i in num_cols:
    plt.subplot(2,2,1)
    sns.histplot(train[i],bins=80,kde=True)
    plt.xlabel('Label')
    plt.ylabel('Count')
    plt.title(f'{i} Histogram')
    plt.subplot(2,2,2)
    plt.boxplot(train[i])
    plt.title(f'{i} Boxplot')
    plt.show()
```

**Outlier Data**

```python
for i in num_cols:
    Q1=np.percentile(train[i],25)
    Q2=np.percentile(train[i],50)
    Q3=np.percentile(train[i],75)
    IQR=Q3-Q1
    lb=Q1-1.5*IQR
    ub=Q3+1.5*IQR
    con1=train[i]<lb
    con2=train[i]>ub
    con=con1|con2
    new_data=np.where(con,Q2,train[i])
    train[i]=new_data
for i in num_cols:
    plt.boxplot(train[i])
    plt.title(i)
    plt.show()
```

**Non-Outlier Data**

```python
for i in num_cols:
    Q1=np.percentile(train[i],25)
    Q2=np.percentile(train[i],50)
    Q3=np.percentile(train[i],75)
    IQR=Q3-Q1
    lb=Q1-1.5*IQR
    ub=Q3+1.5*IQR
    con1=train[i]>lb
    con2=train[i]<ub
    new_data = np.where(con1, lb, train[i])
    train[i] = new_data
    new_data = np.where(con2, ub, train[i])
    train[i] = new_data
for i in num_cols:
    plt.boxplot(train[i])
    plt.title(i)
    plt.show()
```

- To convert the we have some encoding techniques
    - Label Encoder
